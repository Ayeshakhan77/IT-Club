### Handling missing values

```python
data.isnull().sum().sort_values(ascending=False)
```

### Filling missing values using imputation techniques

```python
# Fill NaN values in 'rating' column with its mean
data['rating'].fillna(data['rating'].mean())
```

```python
data['review_language'] = data['review_language'].fillna(data['review_language'].mode()[0])
```

### Handling Outliers

```python
# Detect and remove outliers from 'rating'
Q1 = data['rating'].quantile(0.25)
Q3 = data['rating'].quantile(0.75)
IQR = Q3 - Q1

# Filter
data = data[~((data['rating'] < (Q1 - 1.5 * IQR)) | (data['rating'] > (Q3 + 1.5 * IQR)))]
```

### Normalization and standardization

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Normalization
scaler = MinMaxScaler()
data[['rating']] = scaler.fit_transform(data[['rating']])

# OR Standardization
scaler = StandardScaler()
data[['rating']] = scaler.fit_transform(data[['rating']])
```

### Preprocessing script

```python
def preprocess_dataset(input_file, output_file):
    # Load dataset
    data = pd.read_csv("multilingual_mobile_app_reviews_2025.csv")

    # Handle missing values
    data['rating'] = data['rating'].fillna(data['rating'].mean())
    data['review_language'] = data['review_language'].fillna(data['review_language'].mode()[0])

    # Remove outliers
    Q1 = data['rating'].quantile(0.25)
    Q3 = data['rating'].quantile(0.75)
    IQR = Q3 - Q1
    data = data[~((data['rating'] < (Q1 - 1.5 * IQR)) | (data['rating'] > (Q3 + 1.5 * IQR)))]

    # Normalize ratings
    scaler = MinMaxScaler()
    data[['rating']] = scaler.fit_transform(data[['rating']])

    # Save cleaned data
    data.to_csv(output_file, index=False)
    print("Preprocessing complete! Cleaned file saved as:", output_file)

# Run script
if __name__ == "__main__":
    preprocess_dataset("data.csv", "cleaned_data.csv")
```
